```python
def add_photo_to_popup(jpeg_photo, lat, lon, width, height, resolution):
    html = '<img src="data:image/jpeg;base64,{}">'.format
    #iframe = folium.element.IFrame(html(jpeg_photo.decode('UTF-8')), width=(width*resolution)+20, height=(height*resolution)+20)
    iframe = folium.element.IFrame(html(jpeg_photo.decode('UTF-8')), width, height)
    popup = folium.Popup(iframe, max_width=2650)
    icon = folium.Icon(color="lightgreen", icon="ok")
    marker = folium.Marker(location=[lat, lon], popup=popup, icon=icon)
    marker.add_to(m)
```

```python
# Note: not using this anymore - using plot_line below

def plot_walk(walk_df, freq=100):
    # mark each station as a point
    
    colour = 'blue'

    count = 0
    for index, row in walk_df.iterrows():
        count+=1
        if count%freq == 0:
            folium.CircleMarker([row['lat'], row['lon']],
                                radius=8,
                                popup = str(row['dist']/1e3),
                                # fill_color="#3db7e4", # divvy color
                                fill_color=colour,
                               ).add_to(m)
```

```python
def plot_line(walk_df, m, is_even, freq=100):
    points = []
    count = 0
    for index, row in walk_df.iterrows():
        count+=1
        if count%freq == 0:
            points.append((row['lat'], row['lon']))
    if is_even:
        folium.PolyLine(points).add_to(m)
    else:
        folium.PolyLine(points, color='red').add_to(m)
```
